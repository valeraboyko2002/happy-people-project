```python
import pandas as pd


df = pd.read_csv('data\\world_happiness_combined.csv',sep=';')
df.head()
```

```python
missing_values = df.isna().sum()
print(missing_values[missing_values > 0])

missing_percent = (df.isna().sum() / len(df)) * 100
print(missing_percent[missing_percent > 0].round(2))

numerical_cols = ['Happiness score', 'GDP per capita', 'Social support',
                  'Healthy life expectancy', 'Freedom to make life choices',
                  'Generosity', 'Perceptions of corruption']

print(df[numerical_cols].describe())

print(f"всего записей: {len(df)}")
print(f"уникальных стран: {df['Country'].nunique()}")
print(f"уникальных годов: {sorted(df['Year'].unique())}")
print(f"уникальных регионов: {df['Regional indicator'].nunique()}")

year_counts = df['Year'].value_counts().sort_index()
print('записи по годам: ', year_counts)
```
