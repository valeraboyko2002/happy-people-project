```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

df = pd.read_csv('data/world_happiness_2024.csv', sep = ';', decimal=',')
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
numeric_columns = ['Ladder score', 'GDP per capita', 'Social support', 
                   'Healthy life expectancy', 'Freedom to make life choices', 
                   'Generosity', 'Perceptions of corruption']
```

# ВВП vs Уровень счастья

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(df['GDP per capita'], df['Ladder score'], alpha=0.6)
plt.xlabel('GDP per capita')
plt.ylabel('Ladder score (Happiness)')
plt.title('GDP vs Happiness Score')
plt.grid(True, alpha=0.3)
```

# 10 счастливых стран (горизонтальные бары)

```python
top_10 = df.nlargest(10, 'Ladder score')

plt.subplot(1, 2, 2)
top_10_sorted = top_10.sort_values('Ladder score')
plt.barh(top_10_sorted['Country'], top_10_sorted['Ladder score'])
plt.xlabel('Ladder score')
plt.title('Top 10 Happiest Countries')
plt.tight_layout()
plt.show()
```

# КОРРЕЛЯЦИЯ С ПОКАЗАТЕЛЕМ СЧАСТЬЯ 


```python
correlation_cols = ['Ladder score', 'GDP per capita', 'Social support',
                    'Healthy life expectancy', 'Freedom to make life choices',
                    'Generosity', 'Perceptions of corruption']

correlations = df[correlation_cols].corr()['Ladder score'].sort_values(ascending=False)
for factor, corr in correlations.items():
    print(f"{factor:30}: {corr:+.3f}")
```

# Анализ выбросов

```python
import matplotlib as plt 

plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
plt.figure(figsize=(14, 8))
df[numeric_columns].boxplot()
plt.title('Ящик с усами', fontsize=16,fontweight='bold',pad=20)
plt.xlabel('переменные')
plt.ylabel('значения')
plt.xticks(rotation=45,ha='right')
plt.grid(True,alpha=.3)
plt.tight_layout()
plt.show()
```
