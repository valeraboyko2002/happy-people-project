```python
out_sum = []

for column in numeric_columns:
    data_series = df[column]

    q1 = data_series.quantile(.25)
    q3 = data_series.quantile(.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    out = data_series[(data_series < lower_bound) | (data_series > upper_bound)]
    out_perc = (len(out)/len(data_series)) * 100

    out_country = df.loc[out.index, 'Country'].tolist() if len(out) > 0 else []

    out_sum.append({
        'Показатель':column,
        'кол-во выбросов':len(out),
        '% выбросов': out_perc,
        'примеры стран': out_country if out_country else []
    })
```

# Подготовка данных к прогнозированию

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df_model = df.copy()

X = df_model.drop(['Ranking','Country','Regional indicator','Ladder score'],axis=1)
y = df_model['Ladder score']

X_train,X_test,y_train,y_test = train_test_split(X,y, train_size=.67,random_state=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

feature_names = X.columns.tolist()
```

*Базовые модели сравнение*

```python
import xgboost
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor

models = {
    'Linear Regression':LinearRegression(),
    'Ridge Regression': Ridge(alpha=1),
    'Lasso Regression': Lasso(selection='random',max_iter=1000),
    'Decision Tree': DecisionTreeRegressor(random_state=1),
    'Random Forest': RandomForestRegressor(n_estimators=100,random_state=1),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100,random_state=1),
    'XGBoost': XGBRegressor(),
    'KNN': KNeighborsRegressor(n_neighbors=5),
    'SVR': SVR(kernel='rbf')  
}

def evalute_model(model, X_train,X_test,y_train,y_test, model_name):
    model.fit(X_train,y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    metrics = {
        "model": model_name,
        "train_rmse":np.sqrt(mean_squared_error(y_train,y_train_pred)),
        "test_rmse":np.sqrt(mean_squared_error(y_test,y_test_pred)),
        'train_mae':np.sqrt(mean_absolute_error(y_train,y_train_pred)),
        'test_mae':np.sqrt(mean_absolute_error(y_test,y_test_pred)),
        'train_r2':r2_score(y_train,y_train_pred),
        'test_r2':r2_score(y_test, y_test_pred)
    }

    return model, metrics, y_test_pred

results = []
predictions = {}

for name,model in models.items():
    print(f'\nобучение: {name}')
    trained_model, metrics, y_pred = evalute_model(model,X_train_scaled,X_test_scaled, y_train,y_test,name)
    results.append(metrics)
    predictions[name] = y_pred

    print(f"   Test R²: {metrics['test_r2']:.4f}")
    print(f"  Test RMSE: {metrics['test_rmse']:.4f}")

results_df = pd.DataFrame(results)
results_df.sort_values('test_r2',ascending=False).to_string(index=False)
```
